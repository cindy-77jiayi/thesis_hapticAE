```python
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 1000  # 1000 Hz
duration = 1.0      # 1 second
t = np.linspace(0, duration, int(sample_rate * duration))

vib1 = 0.5 * np.sin(2 * np.pi * 120 * t)
vib2 = np.zeros_like(t)
vib2[200:300] = 1.0
envelope = np.exp(-5 * t)
vib3 = envelope * np.sin(2 * np.pi * 180 * t)

plt.figure(figsize=(10, 4))
plt.plot(t, vib1, label="sine vibration")
plt.plot(t, vib2, label="pulse vibration")
plt.plot(t, vib3, label="decaying vibration")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Synthetic Vibrotactile Signals")
plt.show()
```
